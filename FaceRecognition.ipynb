```python
import cv2
import os
import time
from deepface import DeepFace

# Dossier qui contient les photos connues
dossier = "visages_connus"

# Ouvrir la caméra
camera = cv2.VideoCapture(0)

nom = "Recherche..."
dernier_test = 0

print("Appuie sur Q pour fermer la caméra")

while True:

    succes, image = camera.read()

    if succes == False:
        print("Erreur avec la caméra")
        break

    # Faire la reconnaissance toutes les 2 secondes
    if time.time() - dernier_test > 2:

        dernier_test = time.time()

        try:
            resultats = DeepFace.find(
                img_path=image,
                db_path=dossier,
                model_name="VGG-Face",
                detector_backend="opencv",
                enforce_detection=False,
                silent=True
            )

            if len(resultats) > 0 and resultats[0].empty == False:

                chemin_photo = resultats[0].iloc[0]["identity"]

                nom = os.path.basename(chemin_photo)
                nom = os.path.splitext(nom)[0]

            else:
                nom = "Inconnu"

        except:
            nom = "Aucun visage"

    # Afficher le nom sur l'image
    cv2.putText(
        image,
        nom,
        (30, 50),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2
    )

    # Afficher la caméra
    cv2.imshow("Reconnaissance faciale", image)

    # Appuyer sur Q pour quitter
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Fermer la caméra
camera.release()
cv2.destroyAllWindows()
```
